```python
import requests as r
import pandas as pd
from  bs4 import BeautifulSoup
from selenium import webdriver
from datetime import datetime as dt
```

```python
loc = r"C:\Users\ADMIN\Downloads\chromedriver-win64\chromedriver.exe"
```

```python
url = r"https://www.meesho.com/search?q=man%20watchs%20title&searchType=manual&searchIdentifier=text_search"
```

```python
cService = webdriver.ChromeService(executable_path=loc)
driver  = webdriver.Chrome(service = cService)
driver.get(url)
```

```python
data = driver.page_source
data
```

```python
driver.close()
```

```python
# parser
soup = BeautifulSoup(data,'html.parser')
soup
```

```python
meesho_data={'watches':[],'prices':[],'actual_prices':[],'off in %':[],'time':[]}
for i in soup.find_all('div',{'class':'NewProductCardstyled__ProductHeaderWrapper-sc-6y2tys-30 gspQJ'}):
    if len(i.text.strip().split('')) ==3:
        name,price,actual_price = i.text.strip().split('')
        meesho_data['watches'].append(name)
        meesho_data['prices'].append(price)
        meesho_data['actual_prices'].append(actual_price.split(' ')[0][:-3])
        meesho_data['off in %'].append(actual_price.split(' ')[0][-3:-1])
        meesho_data['time'].append(dt.now())
        # print(actual_price.split(' ')[0][:-3],off)
        # print(actual_price.split(' ')[0][:-3])
    
    # print()
```
